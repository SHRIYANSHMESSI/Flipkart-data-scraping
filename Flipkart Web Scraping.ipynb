```python
from bs4 import BeautifulSoup as bt
```

```python
import requests
```

```python
url1 = "https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_4_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_4_na_na_na&as-pos=2&as-type=RECENT&suggestionId=laptop%7CLaptops&requestId=2bd39c54-169e-4ac2-9401-19cb9e815a9d&as-backfill=on" 
```

```python
resp1=requests.get(url1)
```

```python
resp1
```

```python
data = resp1.text
data
```

```python
resp1.content
```

```python
soup = bt(data, 'lxml')
```

```python
soup
```

```python
soup.find_all()
```

```python
name = []
for i in soup.find_all("div", class_="_4rR01T"):
    name.append(i.text.strip().split("-")[0])
name
```

```python
rate = []
for i in soup.find_all("div", class_="_30jeq3 _1_WHN1"):
    rate.append(i.text.strip())
len(rate)
```

```python
rating = []
for i in soup.find_all("div", class_="_3LWZlK"):
    rating.append(i.text.strip())
rating[0:24]
```

```python
image = []
for i in soup.find_all("img", class_="_396cs4"):
    image.append(i.get("src"))
image[17]
```

```python
import pandas as pd
flipkart = pd.DataFrame({
    "Name" : name,
    "Rate" : rate,
    "Image" : image[0:24],
    "Rating" : rating[0:24]
    })
```
